```python
#Dataset link: https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

#Dataset Description: The dataset contains transactions made by credit cards in September 2013 by European cardholders.

import pandas as pd
df = pd.read_csv('drive/My Drive/creditcard.csv')
print("Successfully Imported Data!")
df.head(6)
```

```python
#Accessing Some Information about Columns DataType
df.info()
```

```python
#identifying NULL values
#ensures that there are no missing values in the dataset.
df.isnull().sum()
```

```python
df.describe()
```

```python
value_counts = df['Class'].value_counts()
print(value_counts)
# here, class refers to whether the transaction wa fraudulent or not
# 0 corresponds to non fraudulent and 1 coprresponds to fraudulent
```

```python
df_0 = df[df['Class']==0]
df_1 = df[df['Class']==1]

df_0_sampled = df_0.sample(n=492,random_state=1)
df_1_sampled = df_1.sample(n=492,random_state=1)

df_balanced = pd.concat([df_0_sampled,df_1_sampled])

# Shuffle the concatenated dataframe
df_balanced = df_balanced.sample(frac=1, random_state=1).reset_index(drop=True)
```

```python
print('Legit Transaction: ', df_0.shape) #Legit Transactions
print('Fraud Transaction: ', df_1.shape) #Fraud Transactions
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
value_counts.plot(kind='bar', color=['Violet', 'Orange'])
plt.title('Distribution of Fraudulent and Non-Fraudulent Transactions')
plt.xlabel('Class')
plt.ylabel('Count')
plt.yscale('log')
plt.xticks(ticks=[0, 1], labels=['Non-Fraudulent=0', 'Fraudulent=1'], rotation=0)
plt.show()
```

```python
# compare the values for both transactions
df.groupby('Class').mean()
```

```python
#Split data into Features and Label
x = df_balanced.drop(['Time','Class'],axis=1)
y = df_balanced.Class
```

```python
x
```

```python
#Split them into training and testing
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state=42)
print('shape of x: ', x.shape)
print('shape of training input: ', x_train.shape)
print('shape of testing input: ', x_test.shape)
```

```python
#Normalization the data
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()

#scaler.fit(x_train,y_train)

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#Model Training
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

#fitting the model
model.fit(x_train,y_train)
```

```python
#Predict on training
predicted_training = model.predict(x_train)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(predicted_training,y_train)

print('Accuracy on Training:',accuracy*100,'%')
```

```python
#Predicting on Testing
predicted_testing = model.predict(x_test)

accuracy = accuracy_score(predicted_testing,y_test)

print('Accuracy on Testing:',accuracy*100,'%')
```

```python
# Get feature importances from the Random Forest model
importances = model.feature_importances_

import numpy as np
# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.title("Feature Importances - Random Forest")
plt.bar(range(len(indices)), importances[indices], align="center")
plt.xticks(range(len(indices)), indices)
plt.xlabel("Feature Index")
plt.ylabel("Importance(Random Forest)")
plt.tight_layout()
plt.show()
```
